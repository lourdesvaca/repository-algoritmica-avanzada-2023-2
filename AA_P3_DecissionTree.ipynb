```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('diamonds.csv')
```

```python
dataset
```

```python
dataset['cut'] = dataset.cut.map({'Ideal':1,'Premium':2, 'Good':3, 'Very Good':4, 'Fair':5})
```

```python
dataset['color'] = dataset.color.map({'E':1,'I':2, 'J':3, 'H':4, 'F':5, 'G':6, 'D':7})
```

```python
dataset['clarity'] = dataset.clarity.map({'SI2':1, 'SI1':2, 'VS1':3, 'VS2':4, 'VVS2':5, 'VVS1':6, 'I1':7, 'IF':8})
```

```python
dataset.describe()
```

```python
corr =  round(dataset.corr(), 4)
corr.style.background_gradient()
```

```python
data_x = dataset.drop(['carat','depth', 'table','price'], axis='columns')
data_y = dataset.price
```

```python
model_tree = DecisionTreeClassifier()
```

```python
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, train_size=0.9)
```

```python
model_tree.fit(train_x, train_y)
```

```python
y_predicted = model_tree.predict(test_x)
```

```python
print(f'Score de prueba: {round(model_tree.score(test_x, test_y), 4) * 100}%')
```

```python
print(f'Score de entrenamiento: {round(model_tree.score(train_x, train_y), 4) * 100}%')
```
